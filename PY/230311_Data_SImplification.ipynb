```python
import pandas as pd
path_derad = r'C:\Git_Repo\Radical_Project\webmap\public\data\DeRadical.geojson'
path_rad = r'C:\Git_Repo\Radical_Project\webmap\public\data\Radical.geojson'
```

# Format  

These are the additional columns in the derad data. 
- Name of Deradicalization Program
- DATE (initiated & discontinued)
- AGENTS
- APPROACH
- SCALE

```python
import geopandas as gpd
df_derad = gpd.read_file(path_derad)
df_rad = gpd.read_file(path_rad)

df_derad = df_derad.drop(['geometry','index_right', 'Type', 'Region'], axis=1)
df_derad = df_derad.rename( columns={
    'Location.1':'Region',
    'DATE (initiated & discontinued)':"Date",
    'Name of Deradicalization Program ': 'Program Name',
    'AGENTS': 'Agents',
    'APPROACH ': 'Approach',
    'SCALE': 'Scale',
    'TARGETS': 'Targets',
    })

df_rad = df_rad.drop(['geometry','index_right', 'Type'], axis=1)

df_rad.columns = [i.strip() for i in df_rad.columns]
df_derad.columns = [i.strip() for i in df_derad.columns]

display( df_derad.head(3) )
display( df_rad.head(3) )
```

```python
# def to create a 4 letter random code
import random
import string
def random_code(code_length=4):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(code_length))

df_rad['uid'] = df_rad['Number'].apply( lambda x: random_code(8) )
df_derad['uid'] = df_derad['Number'].apply( lambda x: random_code(8) )

display( df_rad.head(3) )
display( df_derad.head(3) )
```

```python
import os

data_folder = r'C:\Git_Repo\Radical_Project\webmap\public\data'

df_rad.to_json(os.path.join(data_folder, 'Radical.json') , orient='records')
df_derad.to_json(os.path.join(data_folder, 'DeRadical.json') , orient='records')
```
