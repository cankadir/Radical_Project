```python

import pandas as pd

radical_path = r'C:\Git_Repo\Radical_Project\DATA\D.RAD WP3 Radicalization_LIST_AMENDED1.xlsx'
df_rad = pd.read_excel(radical_path, sheet_name='Sheet1')

deradical_path = r'C:\Git_Repo\Radical_Project\DATA\D.RAD WP3 De-Radicalization_LIST_v2.xlsx'
df_derad = pd.read_excel(deradical_path, sheet_name='Sheet1')

display( df_rad.head() )
display( df_derad.head() )
```

```python
df_rad['Links'] = df_rad['Links'].fillna( df_rad['Link'] )
df_derad['Links'] = df_derad['Links'].fillna( df_derad['Link'] )

df_rad = df_rad.drop( ['lat','lon','quality','confidence',"Link"] , axis = 1 )
df_derad = df_derad.drop( ['lat','lon','quality','confidence','Link'] , axis = 1 )

df_rad = df_rad.dropna( axis=1, how='all' )
df_derad = df_derad.dropna( axis=1, how='all' )

display( df_rad.head() )
display( df_derad.head() )
```

```python
import geocoder

def get_lat_lon( address ):
    g = geocoder.arcgis( address )
    return g.latlng

df_rad['latlon'] = df_rad['address'].apply( get_lat_lon )
df_rad['lat'] = df_rad['latlon'].str[0]
df_rad['lon'] = df_rad['latlon'].str[1]

df_derad['latlon'] = df_derad['address'].apply( get_lat_lon )
df_derad['lat'] = df_derad['latlon'].str[0]
df_derad['lon'] = df_derad['latlon'].str[1]

display( df_rad.head() )
display( df_derad.head() )
```

```python
pd.set_option( 'display.max_columns' , None )
df_rad1 = df_rad.drop( ['latlon'] , axis = 1 )
df_derad1 = df_derad.drop( ['latlon'] , axis = 1 )

import geopandas as gpd

geo_rad1 = gpd.GeoDataFrame( df_rad1, geometry=gpd.points_from_xy( df_rad1.lon, df_rad1.lat ), crs=4326 )
geo_derad1 = gpd.GeoDataFrame( df_derad1, geometry=gpd.points_from_xy( df_derad1.lon, df_derad1.lat ), crs=4326 )

world = gpd.read_file( r'C:\Git_Repo\Radical_Project\DATA\World_SHP\ne_50m_admin_0_countries.shp' )
world = world[['geometry','NAME']]
world.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots( figsize=(10, 6))

geo_rad1.plot( ax=ax, facecolor='r', edgecolor='none' , zorder=1 , markersize=5)
geo_derad1.plot( ax=ax, facecolor='g', edgecolor='none' , zorder=1 , markersize=5)
world.plot( ax=ax, facecolor='grey', edgecolor='w' , zorder=0 , lw = 0.01)
```
