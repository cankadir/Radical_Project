#### DATA 93
t_83, t_rad, t_drad

```python
ar_list_93 = pd.read_excel( r"C:\Users\cansu\OneDrive\Desktop\dred_data_2\AR_Data\D9.3 Data rev.xlsx")
ar_list_93.columns = [i.split("(")[0] for i in ar_list_93.columns] 
ar_list_93.columns = ar_list_93.columns.str.strip()

ar_list_93.columns = ar_list_93.columns.str.replace( "Address", 'address' )
ar_list_93['Number'] = [ int(i) for i in range( ar_list_93.shape[0] ) ]

ar_list_93 = ar_list_93[['Number','Name of Deradicalization Program','Country','Region','address','Event Description']]
ar_list_93.columns = [ f"{i}_AR" for i in ar_list_93.columns]

print( ar_list_93.shape[0] , "rows" )
ar_list_93.head(3)
```

### Put them together

```python
df_ar = pd.concat( [t_drad, t_rad, t_83, t_93] )

print( df_ar.shape[0] , "<->", df.shape[0] , "rows" )
df_ar = df_ar.drop( ['address_AR','Name of Deradicalization Program_AR','Number','Location.1_AR'], axis=1 )
df_ar
```

```python
# if any of the  df_ar['Program Name'].values have ‘ as first character and ’ as last char remove it
df_ar['Program Name'] = df_ar['Program Name'].str.strip('‘’').values
df_ar['Program Name'] = df_ar['Program Name'].str.strip('\n').values

df_ar['Program Name'].values
```

```python
df_ar['Region'] = df_ar['Region'].str.replace( "Helsinki" , "Uusimaa" ) 
```
