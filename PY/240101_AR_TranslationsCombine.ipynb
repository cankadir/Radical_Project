## Correct region names


```python
region_replace = {
    "Beograd": "Belgrade District",
    "City Municipality of Ljubljana" : "Ljubljana",
    "Glasgow City" : "Glasgow",
    "Greater Manchester" : "Manchester",
    "Jerusalem District" : "Jerusalem",
    "Tblisi" : "Tbilisi",
    "Tel Aviv District" : "Tel Aviv",
    "West Pomeranian" : "West Pomeranian Voivodeship",
}

print( df['Region'].nunique() )
df['Region'] = df['Region'].str.strip()
# Iterate over each row and replace complete region name with the one from the dictionary
for index, row in df.iterrows():
    if row['Region'] in region_replace:
        df.loc[index, 'Region'] = region_replace[row['Region']]

print( df['Region'].nunique() )
```
